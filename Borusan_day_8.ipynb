```python
dt['fuel_label'] = dt["fuel"].map({
    "Petrol":"Benzin",
    "Diesel":"Dizel",
    "CNG":"SÄ±kÄ±ÅtÄ±rÄ±lmÄ±ÅGaz",
    "LPG":"LPG"})
dt["fuel_label"].head()
```

```python
dtf= pd.DataFrame({
    "car":["bmw","audi"],
    "features":[["Sunroof","leather"],["bluetooth","gps"]]
    })
print(dtf.explode("features"))
```

```python
dt_melt = dt.melt(
    id_vars=["name","year"], # sabit kalacak sÃ¼tunlar
    value_vars=["selling_price","km_driven"], #long'a Ã§evrilecek sÃ¼tunlar
    var_name="metric", value_name="value"

)

print(dt_melt.head())
```
